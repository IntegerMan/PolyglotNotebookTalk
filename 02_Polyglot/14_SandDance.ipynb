```polyglot-notebook
using System.IO;
using Microsoft.Data.Analysis;
using Microsoft.ML;

string contents = File.ReadAllText("../01_Jupyter/Titanic_Processed.csv");

DataFrame ratings = DataFrame.LoadCsvFromString(contents);
ratings.Head(3)
```

```polyglot-notebook
ratings.GetType()
```

```polyglot-notebook
// Get all column names
string[] columnNames = ratings.Columns.Select(c => c.Name).ToArray();
columnNames
```

```polyglot-notebook
// Remove Columns I don't care about
foreach (string columnName in columnNames) {
    if (columnName.StartsWith("is_") || columnName.StartsWith("Has ") || columnName.StartsWith("Aired_")) {
        ratings.Columns.Remove(columnName);
    }
}

ratings.Head(3)
```

```polyglot-notebook
using Microsoft.DotNet.Interactive.Formatting.TabularData;

TabularDataResource tabular = ratings.ToTabularDataResource();
```

```polyglot-notebook
public class Person {
    public string First {get; set;} = "Janelle";
    public string Last {get; set;} = "Doe";
    public string FullName => $"{First} {Last}";
    public string Phone {get; set;} = "555-1234";
    public string Email {get; set;} = "NotReal@SpamAlot.com";
    public override string ToString() => FullName;
}
```

```polyglot-notebook
#r "..\DLLs\Bogus.dll"
```

```polyglot-notebook
using Bogus;

var faker = new Faker<Person>()
    .RuleFor(p => p.First, (f, p) => f.Name.FirstName())
    .RuleFor(p => p.Last, (f, p) => f.Name.LastName())
    .RuleFor(p => p.Phone, (f, p) => f.Phone.PhoneNumber())
    .RuleFor(p => p.Email, (f, p) => f.Internet.Email(p.First, p.Last));

List<Person> people = new();
for (int i = 0; i < 25; i++) {
    people.Add(faker.Generate());
}
people
```

```polyglot-notebook
#r "nuget:Microsoft.DotNet.Interactive.Formatting"
```

```polyglot-notebook
using Microsoft.DotNet.Interactive.Formatting.TabularData;

TableSchema schema = new TableSchema {
    Fields = new TableDataFieldDescriptors
    {
        new("First", TableSchemaFieldType.String),
        new("Last", TableSchemaFieldType.String),
        new("Phone", TableSchemaFieldType.String),
        new("Email", TableSchemaFieldType.String),
    }
};
schema
```

```polyglot-notebook
var peopleData = people.Select(p => new Dictionary<string, object>() {
    ["First"] = p.First,
    ["Last"] = p.Last,
    ["Email"] = p.Email,
    ["Phone"] = p.Phone,
});

peopleData
```

```polyglot-notebook
var tabularDataResource = new TabularDataResource(schema, peopleData);

tabularDataResource
```

```polyglot-notebook
using SandDance.InteractiveExtension;

// Visualize the resulting data using SandDance
SandDanceDataExplorer explorer = tabular.ExploreWithSandDance();
explorer.Display()
```
